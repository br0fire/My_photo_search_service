## 1. Recognition model: ArcFace; Detection models: Yolov8

### Forward photo

```python
import os
from shutil import copyfile
results = DeepFace.find(img_path = "/content/IMG_20240614_220755_918.jpg",
                        db_path = "/content/test/0",
                        detector_backend="yolov8",
                        align=False,
                        enforce_detection = False,
                        model_name = 'ArcFace')[0]

similar_folder_name = 'forward0'
os.makedirs(similar_folder_name, exist_ok=True)

for index, row in results.iterrows():
    file_path = row["identity"]
    file_name = os.path.basename(file_path)
    new_file_path = os.path.join(similar_folder_name, file_name)
    copyfile(file_path, new_file_path)
```

```python
results = DeepFace.find(img_path = "/content/IMG_20240614_220755_918.jpg",
                        db_path = "/content/test/1",
                        detector_backend="yolov8",
                        align=False,
                        enforce_detection = False,
                        model_name = 'ArcFace')[0]

similar_folder_name = 'forward1'
os.makedirs(similar_folder_name, exist_ok=True)

for index, row in results.iterrows():
    file_path = row["identity"]
    file_name = os.path.basename(file_path)
    new_file_path = os.path.join(similar_folder_name, file_name)
    copyfile(file_path, new_file_path)
```

```python
file_count = len([f for f in os.listdir('forward0') if os.path.isfile(os.path.join('forward0', f))])
print(f"Number of False positive': {file_count}")

file_count = len([f for f in os.listdir('forward1') if os.path.isfile(os.path.join('forward1', f))])
print(f"Number of True positive': {file_count}")
```
